```python
cylinders = set(d['cylinders'] for d in mpg)
cylinders
```

```python
ClassbyCylin = []

for c in cylinders:
    sumclass = 0
    cylintypect = 0
    for d in mpg:
        if d['cylinders'] == c:
            sumclass += float(d['class'])
            cylintypect +=1
    ClassbyCylin.append((c, sumclass / cylintypect))
ClassbyCylin.sort(key = lambda x: x[0])
ClassbyCylin
```

```python
model = set(d['model'] for d in mpg)
model
```

```python
classbymodel = []

for m in model:
    summodel = 0
    classmodelct = 0
    for d in mpg:
        if d['model'] == m:
            summodel += float(d['class'])
            classmodelct += 1
    classbymodel.append((m, summodel / classmodelct))
classbymodel.sort(key = lambda x: x[0])
classbymodel
```

# SCIKIT LEARN

```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
digits = datasets.load_digits()
```

```python
print(digits.data)
```

```python
digits.target
```

```python
digits.images[0]
```

```python
from sklearn import svm
```

```python
from sklearn import metrics
```

```python
plt.figure()
images_and_labels = list(zip(digits.images, digits.target))
for index, (image, label) in enumerate(images_and_labels[:4]):
    plt.subplot(2,4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap = plt.cm.gray_r, interpolation = 'nearest')
    plt.title("Training: %i" %label)
```

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
```

```python
classifier = svm.SVC(gamma = 0.001)
```

```python
classifier.fit(data[:n_samples // 2], digits.target[:n_samples // 2])
```

```python
expected = digits.target[n_samples // 2:]
predicted = classifier.predict(data[n_samples // 2:])
```

```python
print("Classification report for classifier %s:\n%s\n"
% (classifier, metrics.classification_report(expected, predicted)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted))
```

```python
images_and_predictions = list(zip(digits.images[n_samples // 2:], predicted))
for index, (image, prediction) in enumerate(images_and_predictions[:4]):
    plt.subplot(2, 4, index + 5)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Prediction: %i' % prediction)
```

```python
plt.figure()
plt.show()
```
